```python
# Import data to create DataFrame
import pandas as pd
df = pd.read_csv("/Users/pornpipatleelakanaweera/Downloads/audible_uncleaned.csv")
df
```

```python
# Check missing value in each column
df.isnull().sum()
```

```python
# Check duplicates value
df.duplicated().sum()
```

```python
# Clean author column
df["author"] = df["author"].str.replace("Writtenby:", "")
df
```

```python
# Split Firstname and Lastname of author
import re
def split_name(name):
	parts = re.findall("[A-Z]{1}[a-z]*", name)
	return " ".join(parts)

df["author"] = df["author"].apply(split_name)
df
```

```python
# Clean narrator column
df["narrator"] = df["narrator"].str.replace("Narratedby:", "")
df
```

```python
# Split Fullname and Lastname of narrator
df["narrator"] = df["narrator"].apply(split_name)
df
```

```python
df['time'].head(10)
```

```python
df['time'].tail(10)
```

```python
# Clean time column
df["time_hr"] = df["time"].str.extract("([0-9]+)\s[hr]", expand = True)
df["time_min"] = df["time"].str.extract("([0-9]+)\s[min]", expand = True)
df
```

```python
# replace null value in time_hr and time_min column
df['time_hr'] = df['time_hr'].fillna(0)
df['time_min'] = df['time_min'].fillna(0)
df
```

```python
# Change Datatype in time_hr and time_min column
df["time_hr"] = df["time_hr"].apply(lambda x: int(x))
df["time_min"] = df["time_min"].apply(lambda x: int(x))
```

```python
# Calculate duration per episode in second unit
df["duration_sec"] = (df["time_hr"] * 60 * 60) + (df["time_min"] * 60)
df
```

```python
# Drop time, time_hr, time_min column out of DataFrame
df.drop(columns = {"time", "time_hr", "time_min"}, inplace = True)
df
```

```python
# Change format of date in releasedate column
df["releasedate"] = pd.to_datetime(df["releasedate"], format = "%d-%m-%y")
df = df.rename(columns = {"releasedate" : "release_date"})
df
```

```python
df['language'].value_counts()
```

```python
# Capitalize data in language column
df["language"] = df["language"].str.capitalize()
df
```

```python
# Clean stars column
# Split value in stars column into 2 column
df[["stars", "raters"]] = df["stars"].str.split("stars", expand = True)
df
```

```python
# remove 5 out of 5 from value in stars column
df["stars"] = df["stars"].str.replace(" out of 5", "")
df
```

```python
# remove ratings and rating from value in raters column
df["raters"] = df["raters"].str.replace(" ratings", "")
df["raters"] = df["raters"].str.replace(" rating", "")
df
```

```python
# drop row that value in stars column equal to Not rated yet
for x in df.index:
	if df.loc[x, "stars"] == "Not rated yet":
		df.drop(x, inplace = True)

df
```

```python
# drop row that value in raters column equal to None
for x in df.index:
	if df.loc[x, "raters"] == "None":
		df.drop(x, inplace = True)

df
```

```python
df['raters'].str.contains(",").sum()
```

```python
# remove comma from value in raters column
df["raters"] = df["raters"].str.replace(",", "")
df
```

```python
# convert datatype of stars column to float and raters column in integer
df["stars"] = df["stars"].apply(lambda x: float(x))
df["raters"] = df["raters"].apply(lambda x: int(x))
```

```python
df["price"].str.contains(",").sum()
```

```python
# Clean price column
# remove comma from value in price column
df["price"] = df["price"].str.replace(",", "")
df
```

```python
# replace "Free" with 0 in price column
df["price"] = df["price"].str.replace("Free", "0")
df
```

```python
# convert datatype of price column to float
df["price"] = df["price"].apply(lambda x: float(x))
df
```

```python
# Reset index of DataFrame
df.reset_index(drop = True, inplace = True)
df
```

```python
# Check missing values again
df.isnull().sum()
```

```python
# Check duplicates value again
df.duplicated().sum()
```

```python
# Export file to csv
df.to_csv("audible_clean.csv")
```
